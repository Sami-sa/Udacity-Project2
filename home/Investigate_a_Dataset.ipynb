For example, people less than 15 year of age would be depending on their parents to schedule and attend the appointment.

```python
# creating age intervals every 15 years, from 0:15:30:...etc, and check result.
age_to = [] 
for value in df_cleaning_age['age']: 
    if value < 15: 
        age_to.append(15) 
    elif value < 30: 
        age_to.append(30)
    elif value < 45: 
        age_to.append(45)
    elif value < 60: 
        age_to.append(60)
    elif value < 75: 
        age_to.append(75)
    elif value < 90: 
        age_to.append(90)
    elif value < 105: 
        age_to.append(105)
    elif value < 120: 
        age_to.append(120) 
 
       
df_cleaning_age['age_to'] = age_to 
clean_df = df_cleaning_age
clean_df.head(20)
```
